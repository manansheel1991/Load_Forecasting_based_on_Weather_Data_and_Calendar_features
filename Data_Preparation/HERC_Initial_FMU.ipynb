```python
list_dfs_2 = []

for i in range(len(list_of_dfs)):
    
    if (i == 0 or i%2 == 0):
        a_df = list_of_dfs[i].append(list_of_dfs[i+1])
        a_df.columns = a_df.iloc[0]
        a_df = a_df[a_df["TIME"]!="TIME"]
        a_df.reset_index(inplace=True, drop=True)
        list_dfs_2.append(a_df)
        
    else:
        pass
```
